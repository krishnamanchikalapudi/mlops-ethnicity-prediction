```python
dataset_path = "./dataset"
dataset_original = "{}/{}".format(dataset_path, 'ethnicity.csv')
dataset_original
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv(dataset_original)
df
```

```python
names_ethnicity = ['White', 'Black', 'Asian', 'Indian', 'Other']
ethnicity_unique = df['ethnicity'].unique()
ethnicity_unique = np.sort(ethnicity_unique)
ethnicity_unique
```

```python
for x in ethnicity_unique:
  print(f'id: {x} value: {names_ethnicity[x]}')
```

```python
def ethnicity_remove(num, filename):
    df_no_ethnicity = df[df.ethnicity !=  num]
    df_no_ethnicity.to_csv(filename)
    print (f"file saved at {filename} ")
```

```python
def ethnicity_unique(df):
    ethnicity_unique = df['ethnicity'].unique()
    ethnicity_unique = np.sort(ethnicity_unique)
    for x in ethnicity_unique:
        print(f'id: {x} value: {names_ethnicity[x]}')
```

## Remove 'white', ethnicity=0

```python
csv_no_white = "{}/{}".format(dataset_path, 'ethnicity_no_white.csv')
ethnicity_remove(0, csv_no_white)

# Load file
no_white_df = pd.read_csv(csv_no_white)
ethnicity_unique(no_white_df)
no_white_df
```

```python
csv_no_black = "{}/{}".format(dataset_path, 'ethnicity_no_black.csv')
ethnicity_remove(1, csv_no_black)

# load file
no_black_df = pd.read_csv(csv_no_black)
ethnicity_unique(no_black_df)
no_black_df
```
